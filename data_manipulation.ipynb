```python
# MERGING DATAFRAMES

# Create dummy data
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']},
                   index = [0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'C': ['C4', 'C5', 'C6', 'C7'],
                     'D': ['D4', 'D5', 'D6', 'D7']},
                   index = [4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                     'B': ['B8', 'B9', 'B10', 'B11'],
                     'C': ['C8', 'C9', 'C10', 'C11'],
                     'D': ['D8', 'D9', 'D10', 'D11']},
                   index = [8, 9, 10, 11])
```

```python
# APPLY FUNCTION

# It is a faster method than just using a for loop over the dataframe. 

# Access row wise
df.apply(lambda x: x)

# Access first row
df.apply(lambda x: x[0])

# Access first column
df.apply(lambda x: x[0], axis = 1)

# Access by column name
df.apply(lambda x: x['Item_Fat_Content'], axis = 1)
```

```python
# Apply a condition
def clip_price(price):
    if price > 200:
        price = 200
    return price

df['Item_MRP'].apply(lambda x: clip_price(x))[:5] # Doesn't change the original dataframe.
```

```python
def label_encode(city):
    if city == 'Tier 1':
        label = 0
    elif city == 'Tier 2':
        label = 1
    else:
        label = 2
    return label

df['Outlet_Location_Type'] = df['Outlet_Location_Type'].apply(label_encode) # Change the original dataframe
df['Outlet_Location_Type'][:5]
```

```python
# AGGREGATING DATA

# Read the dataset
df = pd.read_csv('bigmart_data.csv')

# Drop the null values
df = df.dropna(how = "any")

# View the top result
df.head()
```

```python
# Group price based on Item_Type
price_by_item = df.groupby('Item_Type')

# Display first rows
price_by_item.first()

# Mean price by item
price_by_item['Item_MRP'].mean()
```

```python
# Group on multiple columns
multiple_columns_group = df[:10].groupby(['Item_Type', 'Item_Fat_Content'])
multiple_columns_group.first()
```

```python
# Generate crosstab of Outlet_Size and Outlet_Location_Type (How are outlet sizes distributed based on the city type?)
pd.crosstab(df['Outlet_Size'], df['Outlet_Location_Type'], margins = True) 
```

```python
# Create pivot table (How are the sales changing per year)
pd.pivot_table(df, index = ['Outlet_Establishment_Year'], values = ['Item_Outlet_Sales'])
```

```python
pd.pivot_table(df, index = ['Outlet_Establishment_Year', 'Outlet_Location_Type', 'Item_Type'], values = ['Item_Outlet_Sales'])
```

```python
pd.pivot_table(df, index = ['Outlet_Establishment_Year', 'Outlet_Location_Type', 'Outlet_Size'], values = ['Item_Outlet_Sales'], aggfunc = [np.mean, np.median, min, max, np.std])
```
