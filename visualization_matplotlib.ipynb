```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Display plots on the notebook itself
%matplotlib inline
```

```python
height = [150, 160, 165, 185]
weight = [70, 80, 90, 100]

# Draw the plot
plt.plot(height, weight)

# Add title
plt.title('Relationship between heigh and weight')

# Label x asis
plt.xlabel('Height')

# Label y axis
plt.ylabel('Weight')
```

```python
calories_burnt = [65, 75, 95, 99]

# Draw the plot for calories burnt
plt.plot(calories_burnt)

# Draw the plot for weight
plt.plot(weight)

# Add legend 
plt.legend(labels = ['Calories Burnt', 'Weight'], loc = 'lower right')

# Set labels for each of these people
plt.xticks(ticks = [0, 1, 2, 3], labels = ['p1', 'p2', 'p3', 'p4'])
```

```python
# Figure size in inches
plt.figure(figsize = [15, 5])

# Draw the plot (g = green, o = dot, y = yellow, -- = --)
plt.plot(calories_burnt, 'go')
plt.plot(weight, 'y--')

# Add legend
plt.legend(labels = ['Calories Burnt', 'Weight'], loc = 'lower right')

# Set labels for each of these people
plt.xticks(ticks = [0, 1, 2, 3], labels = ['p1', 'p2', 'p3', 'p4'])
```

```python
# Create two plots 
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = [6, 6])

# Plot on row 0 and column 0
ax[0, 0].plot(calories_burnt, 'go')

# Plot on row 0 and column 1
ax[0, 1].plot(weight)

# Set titles for subplots
ax[0, 0].set_title('Calories Burnt')
ax[0, 1].set_title('Weight')

# Set ticks for each of these people
ax[0, 0].set_xticks(ticks = [0, 1, 2, 3])
ax[0, 1].set_xticks(ticks = [0, 1, 2, 3])

# Set labels for each of these people
ax[0, 0].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4'])
ax[0, 1].set_xticklabels(labels = ['p1', 'p2', 'p3', 'p4'])
```

```python
# Read the dataset
df = pd.read_csv('bigmart_data.csv')
df = df.dropna(how = 'any')
# df.head()
```

```python
# LINE CHART

price_by_item = df.groupby('Item_Type').Item_MRP.mean()[:10]

x = price_by_item.index.tolist()
y = price_by_item.values.tolist()

# Set figure size
plt.figure(figsize = (16, 8))

# Set title 
plt.title('Mean price for each item type')

# Set axis labels
plt.xlabel('Item Type')
plt.ylabel('Mean Price')

# Set xticks
plt.xticks(labels = x, ticks = np.arange(len(x)))

plt.plot(x, y)
```

```python
# BAR CHART

# Sales by Outlet Size 
sales_by_outlet_size = df.groupby('Outlet_Size').Item_Outlet_Sales.mean()

# Sort by Sales
sales_by_outlet_size.sort_values(inplace = True)

x = sales_by_outlet_size.index.tolist()
y = sales_by_outlet_size.values.tolist()

# Set title 
plt.title('Mean sales for each outlet size')

# Set axis labels
plt.xlabel('Outlet Size')
plt.ylabel('Sales')

# Set xticks
plt.xticks(labels = x, ticks = np.arange(len(x)))

plt.bar(x, y, color = ['red', 'orange', 'magenta'])
```

```python
# HISTOGRAM

# Set title
plt.title('Item MRP (price) distribution')

# Set axis labels
plt.xlabel('Item_MRP')
plt.ylabel('Frequency')

# Plot histogram
plt.hist(df['Item_MRP'], bins = 20, color = 'lightblue')
```

```python
# BOX PLOTS

data = df[['Item_Outlet_Sales']]

# Create outlier point shape
red_diamond = dict(markerfacecolor = 'r', marker = 'D')

# Set title
plt.title('Item Sales distribution')

# Make the box plot
plt.boxplot(data.values, labels = ['Item Sales'], flierprops = red_diamond)
```

```python
data = df[['Item_Weight', 'Item_MRP']]

# Create outlier point shape
red_diamond = dict(markerfacecolor = 'r', marker = 'D')

# Generate subplots
fig, ax = plt.subplots()

# Make the box plot
plt.boxplot(data.values, labels = ['Item Wight', 'Item MRP (price)'])
```

```python
# VIOLIN PLOTS

data = df[['Item_Weight', 'Item_MRP']]

# Generate suplots
fig, ax = plt.subplots()

# Add labels to x axis
plt.xticks(ticks = [1, 2], labels = ['Item_Weight', 'Item_MRP'])

# Make the box plot
plt.violinplot(data.values)
```

```python
# SCATTER PLOTS

# Set label of axis
plt.xlabel('Item_Weight')
plt.ylabel('Item_Visibility')

# Plot
plt.scatter(df['Item_Weight'][:200], df['Item_Visibility'][:200])
```

```python
# BUBBLE PLOTS

# Set label of axis
plt.xlabel('Item_MRP')
plt.ylabel('Item_Outlet_Sales')

# Set title
plt.title('Item Outlet Sales vs Item MRP (price)')

# Plot
plt.scatter(df['Item_MRP'][:100], df['Item_Outlet_Sales'][:100], s = df['Item_Visibility'][:100]*1000, color = 'red')
```
